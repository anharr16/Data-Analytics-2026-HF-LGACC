```python
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
df['Percent Scoring 65 or Above'] = pd.to_numeric(df['Percent Scoring 65 or Above'], errors='coerce')
```

```python
df = df.dropna(subset=['Percent Scoring 65 or Above'])
```

```python
df.head()
```

```python
df[(df['School DBN'] == "27Q323") & (df['Regents Exam'] == "Integrated Algebra")]
```

```python
scholars_dbn = "27Q323"
```

```python
scholars_df = df[df['School DBN'] == scholars_dbn]
```

```python
scholars_df.head()
```

```python
queens_df = df[df['School DBN'].str.contains('Q')]
```

```python
queens_df.head()
```

```python
queens_df.describe()
```

```python
scholars_df['Percent Scoring 65 or Above'].describe()
```

```python
queens_df['Percent Scoring 65 or Above'].describe()
```

```python
print(queens_df['Percent Scoring 65 or Above'].mean())
```

```python
queens_df['Percent Scoring 65 or Above'].plot(kind='hist', bins=20,)
```

```python
queens_avg = queens_df['Percent Scoring 65 or Above'].mean()
```

```python
print(scholars_df['Percent Scoring 65 or Above'].mean())
```

```python
scholars_df['Percent Scoring 65 or Above'].plot(kind='hist', bins=20)
```

```python
scholars_avg = scholars_df['Percent Scoring 65 or Above'].mean()
```

```python
df = pd.DataFrame({
    'Category': ['School', 'Queens'],
    'Percent': [scholars_avg, queens_avg]
}).set_index('Category')
```

```python
df['Percent'].plot(
    kind='bar',
    color=['turquoise', 'gold'],
    title="Average % Scoring 65+"
)
```

```python
scholars_year_avg = scholars_df.groupby('Year')['Percent Scoring 65 or Above'].mean()
```

```python
print(scholars_year_avg)
```

```python
queens_year_avg = queens_df.groupby('Year')['Percent Scoring 65 or Above'].mean()
```

```python
print(queens_year_avg)
```

```python
df = pd.DataFrame({
    'School': scholars_year_avg,
    'Queens': queens_year_avg
})

df.plot(
    kind='bar',
    color=['turquoise', 'gold'],
    title="Average % Scoring 65+ by Year"
)
```

According to the data Scholars Academy has consistently performed well with passing Exam scores averaging 94.76% passing rate. Scholars outperforms its borough (Queens) which averages a 64.96% passing rate. This performance is consistent accross the 3 years of exam data available.
