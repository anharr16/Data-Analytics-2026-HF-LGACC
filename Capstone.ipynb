Grouping age range into age bands to have to see a high level of age impact across ranges vs each individual age

```python
df['Age_Group']=pd.cut(df['Age'],bins=[18,30,40,50,65], labels=['18-30','31-40','41-50','51-65'])
```

```python
df['Age_Group'].head()
```

```python
df['Remote_Work_Frequency'].unique()
```

Assiging label identifiers to remote figures to align with frequency

```python
remote_map = {0:'Fully On Sight', 25: 'Mostly On Sight', 50: 'Hybrid', 75: 'Mostly Remote', 100:'Fully Remote'}
df['Remote_Work_Label'] = df['Remote_Work_Frequency'].map(remote_map)
```
